```python
# Corr heatmap

plt.figure (figsize= (10,10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, cmap = 'coolwarm', annot =True, square = True, fmt = '.2f')
plt.title ('Corr heatmap')

plt.show()
```

```python
# + Proportional, - Inversely Proportional
# i) Temperature and casual users: 0.47. As the temperature rises, more people visit.
# ii) Humidity and casual users: -0.35. As the number of visitors increases, humidity tends to decrease.

# iii) Working day and casual users: -0.32. Fewer working days mean more visitors! However, for registered users, the correlation is 0.12, which is proportional.
# This indicates that registered users tend to book more on weekdays rather than weekends!

# iv) Season/weather factors: Elements like season and wind speed, which influence the weather, also show a similar positive correlation with rental frequency.

# Personally, I also prefer going swimming on weekdays rather than weekends because I assume there will be fewer people!
```

```python
from scipy.stats import ttest_ind

def t_tester (val1, val2):
    t_stat, p_val = ttest_ind(val1, val2)
    if p_val <= 0.5:
        print("p-value =", round(p_val, 3),"reject the null hypothesis.")
    else:
        print("p-value =", round(p_val, 3)," \n does not reject the null hypothesis.")

t_tester (df['temp'], df['casual'])
t_tester (df['humidity'], df['casual'])
t_tester (df['workingday'], df['casual'])
```

```python
# Variables with a correlation value around 0.3 have a weak correlation, meaning a t-test would show only a slight relationship.
# As explained, the significance of temp, working day, and humidity was also confirmed through the t-test.
```

# REAL ML without overfitting

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

x_val = df.drop(columns='datetime')

# Create dummy variables
train_df = pd.get_dummies(x_val, columns=['year', 'month', 'hour', 'holiday', 'workingday', 'season', 'weather'])
test_df = pd.get_dummies(x_val, columns=['year', 'month', 'hour', 'holiday', 'workingday', 'season', 'weather'])

# Align the two DataFrames to match the columns
train_df, test_df = train_df.align(test_df, join='left', axis=1)
test_df = test_df.drop(['count'], axis=1)

# Split features (X) and target (y)
X = train_df.drop(['count'], axis=1)
y = train_df['count']

# Standardize the data using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Ridge regression model (L2 regularization to address overfitting)
ridge_reg = Ridge(alpha=1.0)

# Evaluate the model performance using cross-validation
cv_scores = cross_val_score(ridge_reg, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
print(f"Cross-validation average MSE: {-cv_scores.mean()}")

# Train the model
ridge_reg.fit(X_train, y_train)

# Make predictions
pred = ridge_reg.predict(X_test)

# R-squared
r2 = r2_score(y_test, pred)
print(f"R-squared: {r2}")

# Root Mean Squared Error (RMSE)
rmse = root_mean_squared_error(y_test, pred)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
